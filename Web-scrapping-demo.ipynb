```python
import requests
import pandas as pd
from pandas import json_normalize

# Fetch the data
response = requests.get('https://newsapi.org/v2/everything?q=tesla&from=2024-07-13&sortBy=publishedAt&apiKey=632615e301d348dbafdabe48acd8e6d7')
data = response.json()

# Check if the response is successful
if data['status'] == 'ok':
    # Normalize the articles data
    articles = data['articles']
    df = json_normalize(articles)
    
    # Display the DataFrame
    print(df.head())
else:
    print("Failed to fetch data:", data['status'])
```
