# Exporting CSV file

```python
df=pd.read_csv('D:/Pandas/Deliveries.csv')
```

```python
temp_df=df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df
```

```python
temp_df.to_csv('batsman_runs.csv',index=False)
```

```python
df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum').to_csv('batsman_vs_temas.csv',index=True)
```

# Export in Excel File


```python
temp_df.to_excel('batsman_runs.xlsx')
```

# Html Exporting


```python
six_heatmp=df.query('batsman_runs==6').pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='count').to_html('no_of_sixes.html')
```
